```python
import pandas as pd 
import json 

with open('./Output/output.txt',  'r', encoding='utf-8') as f:
    content = f.read()   # 返回整个文本内容的字符串 
print(content)

result = content.split("\'")
data = result[1::2]

data_df = pd.DataFrame()
for item in data:
    # print(item)
    s = json.loads(item) 
    converted_data = {d["key"]: d["value"] for d in s}
    df = pd.DataFrame([converted_data])
    data_df = pd.concat([data_df, df], ignore_index=True)
    # print(data_df)

data_df.to_excel("output.xlsx", index=False)
```
